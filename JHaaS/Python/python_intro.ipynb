## Functions

```python
def addLineNumbers(fp_input, fp_output="data/default.addline"):
    linenumbers = 0
    chars = 0
    with open(fp_output, 'w') as f2:
        with open(fp_input) as f:
            lines = f.readlines()
            for i, line in enumerate(lines):
                if 'y' in line.lower():
                    f2.write("line %2i with %03i chars '%s'\n" % (
                        i+1, 
                        len(line.strip()), 
                        line.strip()))
                    linenumbers += 1
                    chars += len(line.strip())
    return {'number lines in file': linenumbers, 
            'number characters in file': chars}
```
